```python
# Importing the necessary libraries
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

# Load data


```python
# Load the Breast Cancer dataset from scikit-learn
data = load_breast_cancer()
data.target[[10, 50, 85]]
# Display dataset target names to understand classification labels
print("Class Labels:", list(data.target_names))
```

```python
# Split the dataset into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=42)
```

```python
# Standardize features to have zero mean and unit variance for better model performance
scaler = StandardScaler()
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)
```

# KNN

```python
# Define and train the K-Nearest Neighbors (KNN) classifier with 5 neighbors
neigh = KNeighborsClassifier(n_neighbors=5)
# Train the model
neigh.fit(X_train_sc, y_train)
# Make predictions on the test dataset
y_predictk = neigh.predict(X_test_sc)
# Generate and display the classification report for model evaluation
print("KNN Classification Report:")
print(classification_report(y_test, y_predictk, target_names=data.target_names))
```

```python
# Compute the confusion matrix to assess misclassifications
cm = confusion_matrix(y_test, y_predictk)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=data.target_names)
disp.plot(cmap=plt.cm.Blues) 
plt.show()
```

# Decision Tree


```python
# Define and train the Decision Tree classifier with default settings
tree = DecisionTreeClassifier(random_state=42)
# Train the model
tree.fit(X_train, y_train)
# Make predictions on the test dataset
y_predict_dt = tree.predict(X_test)
# Generate and display the classification report for model evaluation
print("Decision Tree Classification Report:")
print(classification_report(y_test, y_predict_dt, target_names=data.target_names))
```

```python
# Compute the confusion matrix to assess misclassifications
cm = confusion_matrix(y_test, y_predict_dt)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=data.target_names)
disp.plot(cmap=plt.cm.Blues) 
plt.show()
```

```python
# Define and train the Decision Tree classifier and experiminting with diffetent parameters tuning
tree2 = DecisionTreeClassifier(random_state=42, max_depth=2)
# Train the model
tree2.fit(X_train, y_train)
# Make predictions on the test dataset
y_predict_dt2 = tree2.predict(X_test)
# Generate and display the classification report for model evaluation
print("Decision Tree Classification Report with max_depth=2:")
print(classification_report(y_test, y_predict_dt2, target_names=data.target_names))
```

```python
cm = confusion_matrix(y_test, y_predict_dt2)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=data.target_names)
disp.plot(cmap=plt.cm.Blues) 
plt.show()
```

# Random Forest


```python
# Define and train the Random Forest classifier with 100 trees
rf_classifier = RandomForestClassifier(n_estimators = 100)
# Train the model
rf_classifier.fit(X_train, y_train) 
# Make predictions on the test dataset
y_predict_rf = rf_classifier.predict(X_test)
# Generate and display the classification report for model evaluation
print("Random Forest Classification Report:")
print(classification_report(y_test, y_predict_rf, target_names=data.target_names))
```

```python
# Compute the confusion matrix to assess misclassifications
cm = confusion_matrix(y_test, y_predict_rf)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=data.target_names)
disp.plot(cmap=plt.cm.Blues) 
plt.show()
```

```python
# Define and train the Random Forest classifier with 100 trees and experiminting with different parameters tuning
rf_classifier2 = RandomForestClassifier(n_estimators = 100, max_depth=5, min_samples_split=8)
# Train the model
rf_classifier2.fit(X_train, y_train)
# Make predictions on the test dataset
y_predict_rf2 = rf_classifier2.predict(X_test)
# Generate and display the classification report for model evaluation
print("Random Forest Classification Report with max_depth=5, min_samples_split=8:")
print(classification_report(y_test, y_predict_rf2, target_names=data.target_names))
```

```python
# Compute the confusion matrix to assess misclassifications
cm = confusion_matrix(y_test, y_predict_rf2)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=data.target_names)
disp.plot(cmap=plt.cm.Blues) 
plt.show()
```

```python
# Define and train the Random Forest classifier with 100 trees and experiminting with different parameters tuning
rf_classifier3 = RandomForestClassifier(n_estimators = 100, max_depth=1, min_samples_split=4)
# Train the model
rf_classifier3.fit(X_train, y_train)
# Make predictions on the test dataset
y_predict_rf3 = rf_classifier3.predict(X_test)
# Generate and display the classification report for model evaluation
print("Random Forest Classification Report with max_depth=1, min_samples_split=4:")
print(classification_report(y_test, y_predict_rf3, target_names=data.target_names))
```

```python
# Compute the confusion matrix to assess misclassifications
cm = confusion_matrix(y_test, y_predict_rf3)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=data.target_names)
disp.plot(cmap=plt.cm.Blues) 
plt.show()
```

# Comparison of Classifier Performance


```python
models = ['Random Forest', 'Decision Tree', 'KNN']
accuracy = [0.96, 0.95, 0.95]
macro_precision = [0.97, 0.94, 0.94]
macro_recall = [0.96, 0.94, 0.94]
macro_f1 = [0.96, 0.94, 0.94]

x = np.arange(len(models)) 
width = 0.2  

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width*1.5, accuracy, width, label='Accuracy')
rects2 = ax.bar(x - width/2, macro_precision, width, label='Macro Precision')
rects3 = ax.bar(x + width/2, macro_recall, width, label='Macro Recall')
rects4 = ax.bar(x + width*1.5, macro_f1, width, label='Macro F1-score')

ax.set_ylabel('Scores')
ax.set_title('Comparison of Classifier Performance')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

plt.ylim(0.90, 1.0)
plt.show()
```
